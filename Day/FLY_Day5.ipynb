```python
# 문제 파악 단계
x = [-3, 31, -11, 4, 0, 22, -2, -5, -25, -14]
y = [-2, 32, -10, 5, 1, 23, -1, -4, -24, -13]

print(x)
print(y)
```

```python
import matplotlib.pyplot as plt

# 데이터 비교 -> 선현관계를 이루고 있다는 것을 파악
plt.plot(x, y, 'o-')
plt.show()
```

```python
# 데이터 전처리 과정
import pandas as pd

# 1차원으로 되어있던 데이터를 2차원 데이터로 만든다.
df = pd.DataFrame({'X':x, 'Y':y})
df
```

```python
# 데이터가 너무 많아지면 데이터 갯수를 확인하기 어렵기 때문에 확인하는 법
df.shape
```

```python
# 모델 학습 단계
X_train = df.loc[:, ['X']] # loc이 생각보다 많이 쓰임 -> 구동방식 제대로 알아두기
y_train = df.loc[:, ['Y']]

print(X_train.shape, y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression # 선형회귀 알고리즘
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.coef_, lr.intercept_ # 기울기와 y절편
```

```python
print("기울기:", lr.coef_[0][0])
print("y절편:", lr.intercept_[0])
```

```python
# 예측 단계
import numpy as np

X_new = np.array(11).reshape(1, 1) # 2차원으로 데이터를 제공해야 한다.
X_new
```

```python
lr.predict(X_new) # 학습을 통해서 하나의 공식을 만들어 낸다.
```

```python
X_test = np.arange(11, 16, 1).reshape(-1, 1)
X_test
```

```python
y_pred = lr.predict(X_test)
y_pred
```

### 붓꽃 데이터 분석

```python
# 데이터 수집 - 라이브러리 준비
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
iris.keys()
```

```python
# 정보를 살펴봤다
print(iris['DESCR'])
```

```python
print("데이터셋 크기:", iris['target'].shape)
```

```python
# target은 데이터가 아니라 정답이다
print("데이터셋 내용:", iris['target'])
```

```python
# iris['data'][]
```

```python
iris['data'].shape
```

```python
# 데이터를 가지고 와서 column에 이름을 붙여버린다.
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
df.head()
```

```python
# 이렇게 직접 입력해줘도 된다.
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal-width']
```

```python
# 열 추가 하는 법
df['Target'] = iris['target']
df.head()
```

```python
df.info()
```

```python
# 데이터의 통계값을 알 수 있음
df.describe() 
```

```python
df.isnull().sum()
```

```python
# 중복 데이터 확인
df.duplicated().sum()
```

```python
# 중복된 데이터 보기
df.loc[df.duplicated()]
```

```python
df.loc[(df.sepal_length==5.8)&(df.petal-width==1.9), :]
```

```python
# 중복된 데이터 버리기
df = df.drop_duplicates()
```

```python
df.corr()
```

```python
# 상관 계수 히트맵
sns.heatmap(data=df.corr(), square=True, annot=True)
plt.show()
```

```python
# 열의 값이 어떻게 모여있는 지 알 수 있다.
df['Target'].value_counts()
```

```python
plt.hist(x='sepal_length', data=df)
plt.show()
```

```python
plt.hist(x='sepal_width', data=df)
plt.show()
```

```python
sns.displot(x='petal_length', data=df, kind='kde')
plt.show()
```

```python
sns.displot(x='sepal_length', hue='Target', kind='kde', data=df)
plt.show()
```

```python
sns.pairplot(df, hue='Target', size=2.5, diag_kind='kde')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_data = df.loc[:, 'sepal_length':'petal-width']
y_data = df.loc[:, 'Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data,
                                                    test_size=0.2,
                                                    shuffle=True, # 테스트와 학습의 데이터가 균일하게 하기 위해서
                                                    random_state=20)
```

```python
# knn 알고리즘
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7)

knn.fit(X_train, y_train)
```

```python
y_knn_pred = knn.predict(X_test)
print("예측값", y_knn_pred[:5])
```

```python
X_test[:5]
```

```python
# 성능 평가
from sklearn.metrics import accuracy_score

knn_acc = accuracy_score(y_test, y_knn_pred)
print("정확도:", knn_acc)
```

### 버섯 식용 유무 판별 웹서비스

```python
import pandas as pandas
import joblib
import pickle
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df = pd.read_csv('/content/sample.csv', header=0, encoding='cp949')
df
```

```python
# 데이터를 label과 data로 분리
label = []
data = []

for row_dinse, row in df.iterrows():
  label.append(row.iloc[0])
  row_data = []
  
  for v in row.iloc[1:]:
    row_data.append(v)
  data.append(row_data)
```

```python
label[:5]
```

```python
data[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, label)
```

```python
X_train
```

```python
y_train
```

```python
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
```

```python
predict = clf.predict(X_test)
predict 
```

```python
X_test
```

```python
ac_score = metrics.accuracy_score(y_test, predict)

print("정확도", ac_score) 
```
