```python
# 필요한 라이브러리 import 하기
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

## Plot() 함수
- 수치 데이터를 설정하면 선 그래프(line graph)를 그린다.
- 그래프를 화면에 표시하기 위해 plt.show() 함수를 호출한다.
- 그래프의 타이틀 X축 Y축의 이름을 설정할 수 있다.

```python
# 제곱 형태로 나타난다.
y = np.arange(0, 10)**2
# x값은 임의로 배치된다.
plt.plot(y)
plt.show() # show 메소드가 없어도 그래프가 출력되긴 하지만 넣어주는 것이 좋다.
```

```python
plt.plot([1, 5, 3, 8])
plt.show()
```

```python
x = np.arange(10, 20, 0.1)
y = x**2
plt.plot(x, y)
plt.rc('font', family='NanumBarunGothic')
plt.title('플롯 그래프')
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()
```

## 여러 개의 그래프 그리기
- 색상은 알아서 지정해준다.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

plt.plot(x, y)
plt.plot(x, z)
plt.show()
```

## 그래프 스타일 지정

```python
# 데이터의 속성에 따라 색상을 바꿔도 눈에 잘 안띄는 경우가 있기 때문에 선의 유형을
# 바꾸는 것이 좋다.
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x + 0.5)
y3 = np.sin(x + 1)

plt.plot(x, y1, 'r--')
plt.plot(x, y2, 'b,')
plt.plot(x, y3, 'g-.')
plt.show()
```

```python
x = np.arange(10)

plt.plot(x, x, 'r.--')
plt.plot(x, x+2, 'bo-')
plt.plot(x, x+4, 'gv-')
plt.show()
```

## 그래프 형상(Figure) 설정

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x+0.5)
y3 = np.sin(x+1)

plt.plot(x, y1, 'r--', label='sinA')
plt.plot(x, y2, 'b.', label='sinB')
plt.plot(x, y3, 'g-.', label='sinC')
# legend는 그래프의 이름을 띄워주는 작은 박스를 의미한다.
# plot에 label로 이름을 설정해주면 설정한 이름이 나타난다.
plt.legend(loc='upper right') # 표시하기 좋은 자리를 자기가 알아서 찾아간다.
plt.show()
```

```python
# lim 축의 범위 설정
x = np.linspace(0, 10, 100)
y1 = np.sin(x)

plt.plot(x, y1, 'r--', label='Sin A')
plt.xlim([2, 8])
plt.ylim([-1, 0.8])
plt.show()
```

```python
# ticks는 눈금 설정
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, 'b-.', linewidth=3)
plt.xticks(np.arange(0, 10, 2))
plt.yticks(np.arange(-1.0, 1.1))
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, linewidth=3)
plt.xticks([0, 5, 10], labels=['Jan', 'Feb', 'Mar'])
plt.yticks([-1, 0, 1], labels=['min', 'ave', 'max'])
plt.show()
```

```python
# 눈금 레이블 회전
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, linewidth=3)
plt.xticks([0, 5, 10], labels=['2022-01-01', '2022-01-02', '2022-01-03'], rotation=90)
plt.yticks([-1, 0, 1], labels=['min', 'ave', 'max'])
plt.show()
```

```python
# 눈금 삭제
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, linewidth=3)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# 색상을 정해주는 방법이 많다.
x = np.arange(10)

plt.plot(x, x, color = "r")
plt.plot(x, x+2, color = "green")
plt.plot(x, x+4, color = "#524FA1")
plt.plot(x, x+6, 'k-')
plt.show()
```

## 그래프 구조
- Figure 객체, Axes 객체, Axis 객체 등으로 구성
- Figure 객체는 그림이 그려지는 전체 영역을 말하고, Axes는 하나의 그래프를 의미한다.
Axis는 그래프를 구성하는 x, y축을 의미한다.
- Figure 객체는 한 개 이상의 Axes 객체를 포함한다.

```python
x = np.arange(1, 30)
y = np.log(x)

plt.figure(figsize = (8, 6))
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(1, 30)
y = np.log(x)

plt.plot(x, y)
fig = plt.gcf()
fig.set_size_inches(8, 6)
plt.show()
```

subplot 값에 따른 그래프 갯수 변화

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r-')

plt.subplot(2, 2, 2)
plt.plot(x, z, 'g--')

plt.subplot(2, 2, 3)
plt.plot(x, z, 'b-.')

plt.show()
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots(2, 2)
fig.set_size_inches(10, 8)
ax[0][0].plot(x, y, 'r-')
ax[0][1].plot(x, z, 'g-')
ax[1][0].plot(x, z, 'b-')

plt.show()
```

```python
plt.style.use('ggplot')
plt.rc('font', family='NanumBarunGothic')

fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot(111)
ax.plot(df.index, df['TOEI ANIMATION'], label='TOEI ANIMATION')
ax.plot(df.index, df['IG Port'], label= 'IG Port')
ax.legend(loc='lower right')
ax.set_title('주가 등락률 2년간 추이')
ax.set_xlabel('년월')
ax.set_ylabel('주가등락률')

plt.show()
```

## 막대 그래프 bar() 함수

```python
x = ['A', 'B', 'C', 'D']
y = [5., 25., 50., 20.]

plt.bar(x, y)
plt.show()
```

```python
x = ['A', 'B', 'C', 'D']
y1 = [5., 25., 50., 20.]
y2 = [8., 20., 25., 6.]

fig, ax = plt.subplots(1, 2)
fig.set_size_inches(12, 5)

ax[0].barh(x, y1, color='#33aaff') # 가로로 데이터를 보여줄 수 있다.
ax[1].bar(x, y2, color=['r', 'g', 'b', 'k'])
plt.show()
```

## 두 개의 축을 가진 그래프 그리기

```python
# volume하고 close 값을 비교한다.
df = pd.read_csv('./sample/4816.csv', index_col=0, parse_dates=['Date'])
df.head()
```

```python
fig = plt.figure(figsize=(10, 4))
ax1 = fig.add_subplot(111)
ax1.plot(df.index, df['Close'], color='b', label='주가')

ax2 = ax1.twinx()
ax2.bar(df.index, df['Volume'], color='g', label='거래총액')

ax1.legend()
ax2.legend(loc=6)

ax1.set_xlabel('년월')
ax1.set_ylabel('주가')
ax1.set_yticks([i * 2000 for i in range(5)]) # 0 ~ 4 총 5번 반복 2000 곱하기
ax2.set_yticks([i * 5000 for i in range(5)])
ax2.set_ylabel('거래총액')

plt.show()
```

```python
df = pd.read_csv('./sample/anime_master.csv', index_col='anime_id')
df.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(df['members'], df['rating'], alpha=0.1)

plt.show()
```

```python
# members 80만 이상인 데이터를 출력하시오!
df.loc[df['members']>800000, ['name', 'members']]
```

```python
# 멤버수가 60만 이상 평점이 8.5 이상인 작품, 이름, 평점
df.loc[(df['members']>600000) & (df['rating']>= 8.5), ['name', 'rating']]
```

```python
types = df['type'].unique()
types
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)

for i in types:
  x = df.loc[df['type']== i, 'members']
  y = df.loc[df['type']==i, 'rating']
  ax.scatter(x, y, alpha=0.5, label=i)

ax.legend(loc='lower right', fontsize=15)  
plt.show()
```
