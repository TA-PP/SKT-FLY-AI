```python
#   -  
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
iris.keys()
```

```python
iris['data'].shape
```

```python
#    column  .
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
df.head()
```

```python
#    .
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal-width']
```

```python
#    
df['Target'] = iris['target']
df.head()
```

```python
df.info()
```

```python
#     
df.describe() 
```

```python
#   
df = df.drop_duplicates()
```

```python
df.corr()
```

```python
#   
sns.heatmap(data=df.corr(), square=True, annot=True)
plt.show()
```

```python
#        .
df['Target'].value_counts()
```

```python
sns.pairplot(df, hue='Target', size=2.5, diag_kind='kde')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_data = df.loc[:, 'sepal_length':'petal-width']
y_data = df.loc[:, 'Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data,
                                                    test_size=0.2,
                                                    shuffle=True, #      
                                                    random_state=20)
```

```python
# knn 
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7)

knn.fit(X_train, y_train)
```

```python
y_knn_pred = knn.predict(X_test)
print("", y_knn_pred[:5])
```

```python
X_test[:5]
```

```python
#  
from sklearn.metrics import accuracy_score

knn_acc = accuracy_score(y_test, y_knn_pred)
print(":", knn_acc)
```

```python
import pandas as pandas
import joblib
import pickle
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df = pd.read_csv('/content/sample.csv', header=0, encoding='cp949')
df
```

```python
#  label data 
label = []
data = []

for row_dinse, row in df.iterrows():
  label.append(row.iloc[0])
  row_data = []
  
  for v in row.iloc[1:]:
    row_data.append(v)
  data.append(row_data)
```

```python
label[:5]
```

```python
data[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, label)
```

```python
X_train
```

```python
y_train
```

```python
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
```

```python
predict = clf.predict(X_test)
predict 
```

```python
X_test
```

```python
ac_score = metrics.accuracy_score(y_test, predict)

print("", ac_score) 
```
